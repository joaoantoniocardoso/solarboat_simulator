# HydrostaticHull Example

This example shows how to load the hydrostatic surrogate and query wetted area, trim, and geometry for an operating point.

```python
import json
from pathlib import Path
import sys

sys.path.append('.')
sys.path.append('../../../')

from hydrostatic_hull import HydrostaticHull
from utils.optimization import load_model_params_from_json
```

```python
hull_input_params = load_model_params_from_json('data/hull_input_params.json')

cog_x = hull_input_params['cog_x']
disp_mass = hull_input_params['disp_mass']
total_area = hull_input_params['total_area']

hull = HydrostaticHull(cog_x=cog_x, disp_mass=disp_mass, total_area=total_area)
hull_params = hull.get_params()
hull_params
```
