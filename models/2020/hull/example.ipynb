```python
from hydrostatic_hull import HydrostaticHull

# Initialize the model
hull = HydrostaticHull()  # Assumes data/meshmagick.feather exists

# Your operating condition
cog_x = 1.976      # m from transom
disp_mass = 293.7  # kg

# Get hydrostatic properties
wet_area = hull.wet_surface_area(cog_x, disp_mass)
trim = hull.trim_angle(cog_x, disp_mass)

print(f"Wetted area: {wet_area:.3f} m²")
print(f"Trim angle: {trim:.3f}°")

# Check if in valid range
if hull.is_in_valid_range(cog_x, disp_mass):
    print("Operating point is within validated range")
else:
    print("Operating point is outside training data range")
```
